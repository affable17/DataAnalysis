```python
df = df_raw.melt(
        id_vars=["국가(대륙)별", "상품군별", "판매유형별"],
          var_name="기간",
           value_name="백만원")#column 이름 지정
```

```python
#lambda라는 익명 함수, 기간을 x변수에 받아 
df["연도"] = df["기간"].map(lambda x : int(x.split(".")[0]))
```

```python
df.head()
```

```python
"2014.1/4".split(".")[1].split("/")[0]    #기간에서 분기만 분리
```

```python
df["분기"] = df["기간"].map(lambda x : int(x.split(".")[1].split("/")[0]))
```

```python
df.head()
```

## 금액을 수치데이터로 표현하기 위해 데이터 타입 변경

```python
# - 문자를 결측치로 변경, float 타입으로 변경하기
# - 는 자료없음을 뜻하기 때문이다. 
df["백만원"]= df["백만원"].replace("-", pd.np.nan).astype(float)    
#pandas 내부의 넘파이를 불러올수 있어

df["백만원"]
```

```python
df.head()
```

## 필요없는 데이터 제거하기

```python
df.info()
```

```python
#합계 데이터는 따로 구할 수 있어 -> 데이터에서 제거 
df = df[(df["국가(대륙)별"] != "합계") & 
         (df["상품군별"] != "합계") ].copy()
df.info() #메모리 줄어듦
```

```python
df.isnull().sum()    #true값만 다 더해 
```

## 의류 및 패션 관련 상품 온라인 쇼핑 해외직접판매액

```python
df_fashion = df[(df["상품군별"]=="의류 및 패션 관련상품") 
                & (df["판매유형별"]=="계")].copy()
df_fashion.head()
```

```python
plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion, x='기간', y='백만원',
            hue='국가(대륙)별')
```

```python
#기간별 금액 데이터 시각화 
df_fashion2 = df[(df["상품군별"]=="의류 및 패션 관련상품") 
                & (df["판매유형별"]!="계")].copy()

plt.figure(figsize=(15,4))
plt.xticks(rotation=30)
sns.lineplot(data=df_fashion2, x='기간', y='백만원',
            hue='판매유형별',
            ci=None)
```

## 데이터 집계하기

```python
#df_fashion.pivot_table?
df_fashion.head(2)
```

```python
result = df_fashion.pivot_table(index='국가(대륙)별',
                      columns="연도",
                      values="백만원",
                      aggfunc="sum") #aggfunc = mean is default
result
```

## 연산결과를 시각적으로 보기

```python
plt.figure(figsize=(10,6))
sns.heatmap(result, cmap="Blues", annot=True,
           fmt=".0f") #소숫점 없이 float형태로 표시 
```
